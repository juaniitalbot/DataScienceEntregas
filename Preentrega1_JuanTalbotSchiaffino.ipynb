```python
from datetime import datetime

# Convierte las fechas de tipo texto a tipo fecha
transaction1997['transaction'] = pd.to_datetime(transaction1997['transaction'])

# Agrega una columna para el mes
transaction1997['month'] = transaction1997['transaction'].dt.month

# Agrega una columna para el año
transaction1997['year'] = transaction1997['transaction'].dt.year

# Agrupa los datos por mes y año y suma las ventas
df_monthly1997 = transaction1997.groupby(['year', 'month'])['quantity'].sum().reset_index()

# Crea una nueva columna con las fechas del primer día de cada mes
df_monthly1997['date'] = pd.to_datetime(df_monthly1997[['year', 'month']].assign(day=1))

# Crea una serie con los nombres de los meses
month_names = df_monthly1997['date'].dt.strftime('%B')

sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))

# Crea el gráfico de línea usando Seaborn
fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(x='date', y='quantity', data=df_monthly1997, ax=ax)

# Configura el formato del eje x
ax.xaxis.set_major_locator(plt.MaxNLocator(len(df_monthly1997)))
ax.set_xticklabels(month_names, rotation=45, ha='right')

# Configura los títulos y etiquetas de los ejes
ax.set_title('Ventas mensuales 1997')
ax.set_xlabel('Mes')
ax.set_ylabel('Cantidad vendida')

# Muestra el gráfico
plt.show()
```

Se puede ver como hay una caida de venta en los meses de febrero y marzo. Tambien se ve que para fin de año hay un aumento de ventas que seguro conecta con un aumento en el proximo a­ño.

```python
# Convierte las fechas de tipo texto a tipo fecha
transaction1998['transaction'] = pd.to_datetime(transaction1998['transaction'])

# Agrega una columna para el mes
transaction1998['month'] = transaction1998['transaction'].dt.month

# Agrega una columna para el año
transaction1998['year'] = transaction1998['transaction'].dt.year

# Agrupa los datos por mes y año y suma las ventas
df_monthly1998 = transaction1998.groupby(['year', 'month'])['quantity'].sum().reset_index()

# Crea una nueva columna con las fechas del primer día de cada mes
df_monthly1998['date'] = pd.to_datetime(df_monthly1998[['year', 'month']].assign(day=1))

# Crea una serie con los nombres de los meses
month_names = df_monthly1998['date'].dt.strftime('%B')

sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))

# Crea el gráfico de línea usando Seaborn
fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(x='date', y='quantity', data=df_monthly1998, ax=ax)

# Configura el formato del eje x
ax.xaxis.set_major_locator(plt.MaxNLocator(len(df_monthly1998)))
ax.set_xticklabels(month_names, rotation=45, ha='right')

# Configura los títulos y etiquetas de los ejes
ax.set_title('Ventas mensuales 1998')
ax.set_xlabel('Mes')
ax.set_ylabel('Cantidad vendida')

# Muestra el gráfico
plt.show()
```

Se puede ver el aumento de ventas con relacion al año anterior y que durante este año las ventas se mantuvieron muy igualadas hasta el mes de octubre que otra vez aumentan significativamente las ventas.

```python
# Convertir la columna 'transaction_date' a tipo datetime
transaction1997['transaction'] = pd.to_datetime(transaction1997['transaction'])
transaction1998['transaction'] = pd.to_datetime(transaction1998['transaction'])

# Crear una lista con los nombres de los meses en orden cronológico
meses_ordenados = ['January', 'February', 'March', 'April', 'May', 'June',
                   'July', 'August', 'September', 'October', 'November', 'December']

# Agrupar las ventas por mes y sumarlas para cada dataset
ventas_por_mes_df1 = transaction1997.groupby(transaction1997['transaction'].dt.strftime('%B'))['quantity'].sum().reindex(meses_ordenados)
ventas_por_mes_df2 = transaction1998.groupby(transaction1998['transaction'].dt.strftime('%B'))['quantity'].sum().reindex(meses_ordenados)

# Crear un DataFrame con las ventas de cada mes para ambos datasets
df = pd.concat([ventas_por_mes_df1, ventas_por_mes_df2], axis=1, keys=['Ventas 1997', 'Ventas 1998'])
df.index.name = 'Mes'

df = df.reset_index()

sns.set(style='whitegrid')
plt.figure(figsize=(10, 6))

ax = sns.barplot(x='Mes', y='value', hue='variable', data=pd.melt(df, ['Mes']))
ax.set_title('Ventas por mes')
ax.set_xlabel('Mes')
ax.set_ylabel('Ventas')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
ax.legend(title='Año')

plt.show()
```

En el grafico de barras de arriba podemos ver una compariva de ventas por mes de los diferentes años y notamos un aumento notable en las ventas de 1998

# **Algoritmos de regresion store_id = 1**



```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

#Extraer características de fecha
X =  pd.to_datetime(transactions_s1_p13['transaction']).astype(int).values.reshape(-1, 1)

y = transactions_s1_p13['quantity'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_pred

import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Convertir las fechas de numérico a formato de fecha
dates_test = pd.to_datetime(X_test.reshape(-1)).tolist()

# Graficar ventas reales
plt.scatter(dates_test, y_test, color='blue', label='Ventas reales')

# Graficar predicciones
plt.plot(dates_test, y_pred, color='red', label='Predicciones')

# Personalizar el eje X con formato de fecha
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())

# Rotar las etiquetas del eje X para una mejor legibilidad
plt.gcf().autofmt_xdate()

# Personalizar el gráfico
plt.xlabel('Fecha de transacción')
plt.ylabel('Cantidad')
plt.title('Predicciones de ventas futuras para store_id = 1 y product_id= 13')
plt.legend()

# Mostrar el gráfico
plt.show()

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mse)

print("MSE:",mean_squared_error(y_test,y_pred))
print("MAE:", mae)
print("r2:", r2)
print("Raíz del error cuadrático medio:", rmse)
```
